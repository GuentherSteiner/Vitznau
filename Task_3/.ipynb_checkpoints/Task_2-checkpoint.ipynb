# Estimating the Best model (if possible)
## Rescaling and prepping the dataset

```python
# we have our dat with all the dependent and independent variables inside
# firstly we rescale our data
dat_2 = dat # keep dat_2 as backup

# initialize scaler
scaler = MinMaxScaler()

# apply scaler
dat = pd.DataFrame(scaler.fit_transform(dat), columns = dat.columns)

# have a look
#print(data.head())


# in a second step we create a dataframe for x variables and y variables (i.e., dependent and independent)
dependent_vars = ['MktmRF', 'SMB', 'HML', 'CMA', 'RMW', 'QMJ', 'MOM', 'iMom', 'VOL', 'BAB', '1overN', 'leading_index']
independent_vars = ['avg_hours', 'infexp', 'initial_claims', 'YFFR10', 'building_permits', 'cons_sent', 'orders_CG', 'orders_cap', 'leading_index']
x_data = dat.drop(dependent_vars, axis = 1)
y_data = dat.drop(independent_vars, axis = 1)

# in a third step we create a train-test split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=42, shuffle = False)
```

## Analysing different independent variables by 'hand'

```python
# calculation for newey west standard errors
# Use rule of thumb to calculate number of lags taken into consideration for newey-west HAC and round up
print(0.75*len(x_train)**(1/3))
```

### MktmRF

```python
# now we will apply the recursive feature eliminiation function to eliminate 4 of our 11 independant variables
lm = LinearRegression()
lm.fit(x_train, y_train.loc[:,['MktmRF']])

rfe = RFE(lm, n_features_to_select=6)
rfe = rfe.fit(x_train, y_train.loc[:,['MktmRF']])

# print the list of the variables selected for our regression
list(zip(x_train.columns,rfe.support_,rfe.ranking_))

# create new x_train with only the variables inside that made it past the RFE
# columns where RFE support is true:
col = x_train.columns[rfe.support_]

# now create the new x_train and x_test
x_train_rfe = x_train[col]
x_test_rfe = x_test[col]
col
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity
x = x_train_rfe
xt = x_test_rfe
y = y_train.loc[:, ["MktmRF"]]
yt = y_test.loc[:, ["MktmRF"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":5}) # Newey West SE with 4 lags
fit1.summary()

# from the regression output we see that orders_cap is not significant
# we'll run a new regression with this excluded
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity
x = x.drop(['orders_cap'], axis = 1)
xt = xt.drop(['orders_cap'], axis = 1)
y = y_train.loc[:, ["MktmRF"]]
yt = y_test.loc[:, ["MktmRF"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":5}) # Newey West SE with 4 lags
fit1.summary()

# from the regression output we see that cons_sent is not significant
# we'll run a new regression with this excluded
```

### MktmRF predicition

```python
# we have now found that infexp is a significant predictor for MktmRF, we try to predict now with our x_test and compare it with our y_test

y_test_predict = fit1.predict(sm.add_constant(xt))

resid = []
mape = []
for i in range(len(y_test_predict)):
    j = 235 + i
    resid.append(yt.iloc[i,0] - y_test_predict[j])
    mape.append((1/len(y_test_predict))*abs(yt.iloc[i,0]-y_test_predict[j])/yt.iloc[i,0])

print(np.sum(mape))
print('Durbin Watson:' , durbin_watson(resid))
# average difference between real value and prediction is 16% --> quite high
```

```python
reg_1 = LinearRegression()
reg_1.fit(y, x)
print(reg_1.intercept_)
print(reg_1.coef_)
print('R2 Train Data:', reg_1.score(y, x))
print('R2 Test Data:', reg_1.score(yt, xt))
```

We get all kinds of funny things --> discard the model
