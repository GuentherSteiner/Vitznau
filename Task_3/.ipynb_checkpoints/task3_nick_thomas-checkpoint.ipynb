```python
# Create cumulative Monthly returns out of Daily Strategy index
data_strat.set_index("Date", inplace=True) # bc formula needs datetimeindex

strat_mtl = data_strat.pct_change().resample("M").agg(lambda x: ((1+x).prod()-1)*100) # pct_change creates ordinary returns, resample Monthly and aggregating with the (1+x) -1 formula to get monthly ordinary returns

# set date as index for the data and concat data and strat_mtl in df
data.set_index("Date", inplace=True) # bc data_strat has datetimeindex it is easier to concat
df = pd.concat([data, strat_mtl], axis=1)
```

```python
# apparently the leading_index is a linear combination of the other indices --> drop it from the regression

data = data.drop(['leading_index'], axis = 1)
#data
```

```python
# we have our df with all the dependent and independent variables inside, we have data with only independent and
# strat_mtl with the dependent variables inside

# in a first step we create a train-test split
x_train, x_test, y_train, y_test = train_test_split(data, strat_mtl, test_size=0.3, random_state=42, shuffle = False)

# Use rule of thumb to calculate number of lags taken into consideration for newey-west HAC and round up
print(0.75*len(x_train)**(1/3))
```

### Carry wEM

```python
# now we will apply the recursive feature eliminiation function to eliminate 3 of our 10 independant variables
lm = LinearRegression()
lm.fit(x_train, y_train.loc[:,['Carry_wEM']])

rfe = RFE(lm, n_features_to_select=7)
rfe = rfe.fit(x_train, y_train)

# print the list of the variables selected for our regression
list(zip(x_train.columns,rfe.support_,rfe.ranking_))

# create new x_train with only the variables inside that made it past the RFE
# columns where RFE support is true:
col = x_train.columns[rfe.support_]

# now create the new x_train and x_test
x_train_rfe = x_train[col]
x_test_rfe = x_test[col]
col
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity
x = x_train_rfe
xt = x_test_rfe
y = y_train.loc[:, ["Carry_wEM"]]
yt = y_test.loc[:, ["Carry_wEM"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":4}) # Newey West SE with 4 lags
fit1.summary()

# from the regression output we see that building_permits, RF, average_hours are all not significant
# we'll run a new regression with these excluded
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity
x = x.drop(['building_permits', 'RF', 'avg_hours'], axis = 1)
xt = xt.drop(['building_permits', 'RF', 'avg_hours'], axis = 1)
y = y_train.loc[:, ["Carry_wEM"]]
yt = y_test.loc[:, ["Carry_wEM"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":4}) # Newey West SE with 4 lags
fit1.summary()

# from the regression output we see that RF is not significant
# we'll run a new regression, excluding this
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity
x = x.drop(['cons_sent'], axis = 1)
xt = xt.drop(['cons_sent'], axis = 1)
y = y_train.loc[:, ["Carry_wEM"]]
yt = y_test.loc[:, ["Carry_wEM"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":4}) # Newey West SE with 4 lags
fit1.summary()

# from the regression output we see that infexp is not significant
# we'll run a new regression, excluding this
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity
x = x.drop(['infexp'], axis = 1)
xt = xt.drop(['infexp'], axis = 1)
y = y_train.loc[:, ["Carry_wEM"]]
yt = y_test.loc[:, ["Carry_wEM"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":4}) # Newey West SE with 4 lags
fit1.summary()

# from the regression output we see that YFFR is not significant
# we'll run a new regression, excluding this
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity
x = x.drop(['YFFR10'], axis = 1)
xt = xt.drop(['YFFR10'], axis = 1)
y = y_train.loc[:, ["Carry_wEM"]]
yt = y_test.loc[:, ["Carry_wEM"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":4}) # Newey West SE with 4 lags
fit1.summary()

# from the regression output we see that initial claims is significant --> try to predict
```

### Carry wEM predicition

```python
# we have now found that initial_claims is a significant predictor for MktmRF, we try to predict now with our x_test and compare it with our y_test

y_test_predict = fit1.predict(sm.add_constant(xt))

resid = []
mape = []
for i in range(len(y_test_predict)):
    j = 117 + i
    resid.append(yt.iloc[i,0] - y_test_predict[j])
    mape.append((1/len(y_test_predict))*abs(yt.iloc[i,0]-y_test_predict[j])/yt.iloc[i,0])

print(np.sum(mape))
print('Durbin Watson:' , durbin_watson(resid))
# average difference between real value and prediction is 12% --> quite high
```

```python
reg_1 = LinearRegression()
reg_1.fit(y, x)
print(reg_1.intercept_)
print(reg_1.coef_)
print('R2 Train Data:', reg_1.score(y, x))
print('R2 Test Data:', reg_1.score(yt, xt))
```

```python
# Regression using Newey West HAC standard errors because of the autocorrelation and heteroscedasticity

x = X_train.loc[:,["infexp"]]
y = y_train.loc[:, ["Carry_G10"]]

xt = X_test.loc[:,["infexp"]]
yt = y_test.loc[:, ["Carry_G10"]]


####### I would go with this approach in presence of autocorrelation in the errors ####### 
model1 = sm.OLS(endog=y, exog=sm.add_constant(x))
fit1 = model1.fit(cov_type="HAC",cov_kwds={"maxlags":4}) # Newey West SE with 4 lags
fit1.summary()
## --> Ich hätti so e regression gmacht mit dere cov_type"HAC" machsch newey west SE und d lags chasch mit de formle usrechne damits d autocorrelation guet userechnet
######################################################################################################

# SKLEARN WORKS BUT HAS NO ROBUST STANDARD ERRORS BUILT INTO ITS FIT() FUNCTION
model1 = LinearRegression().fit(x, y)
yt_pred = model1.predict(xt)
resid1 = yt - yt_pred
print("Durbin Watson test statistics is: ",durbin_watson(resid1), "must be between 1.5 and 2.5, else autocorrelation present") # H0: No correlation among residuals -> rejected if test is not between 1.5 and 2.5
print("The model score in sample is: ", model1.score(x, y, sample_weight=None)) # score uses R^2 and best is 1 and 0 means predicting same value always -> negative is shite
print("The model score out of sample is: ", model1.score(xt, yt, sample_weight=None))
```

# Cross Validation

```python
class BlockingTimeSeriesSplit():
    def __init__(self, n_splits):
        self.n_splits = n_splits
    
    def get_n_splits(self, X, y, groups):
        return self.n_splits
    
    def split(self, X, y=None, groups=None):
        n_samples = len(X)
        k_fold_size = n_samples // self.n_splits
        indices = np.arange(n_samples)

        margin = 0
        for i in range(self.n_splits):
            start = i * k_fold_size
            stop = start + k_fold_size
            mid = int(0.8 * (stop - start)) + start
            yield indices[start: mid], indices[mid + margin: stop]
```

```python
btscv = BlockingTimeSeriesSplit(n_splits=5) # ob de shit funktioniert bini mer ned ganz sicher
```
